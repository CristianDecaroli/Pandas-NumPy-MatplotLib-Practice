```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(170, 10, 200)
print(data)
```

```python
plt.hist(data, color='salmon', bins=5, edgecolor='black', alpha=0.6)
plt.title('Distribuci√≥n de alturas')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad')
plt.show()
```

#### Boxplot

Este gr√°fico proporciona una representaci√≥n visual clara de la distribuci√≥n de un conjunto de datos, destacando la mediana, los cuartiles y los valores at√≠picos.

A continuaci√≥n, exploraremos en detalle qu√© es un boxplot, sus componentes principales y c√≥mo puede ser utilizado para obtener informaci√≥n valiosa sobre la dispersi√≥n y asimetr√≠a de los datos.



#### ¬øQu√© es un Boxplot?
Un boxplot, o diagrama de caja, es una herramienta visual utilizada en el an√°lisis exploratorio de datos para resumir la distribuci√≥n de un conjunto de datos. Este gr√°fico representa la mediana, los cuartiles y los posibles valores at√≠picos, proporcionando una visi√≥n r√°pida de la dispersi√≥n y la asimetr√≠a de los datos.

Componentes de un Boxplot:
- **Box (Caja)**: Representa el rango intercuartil (IQR), que incluye el 50% central de los datos. Los bordes de la caja indican el primer cuartil (Q1) y el tercer cuartil (Q3).
-  **Median (Mediana)**: Indica el valor central de los datos, dividiendo el conjunto en dos partes iguales.
-  **Whiskers (Bigotes)**: Extienden desde los cuartiles hasta los valores m√≠nimos y m√°ximos dentro de 1.5 veces el IQR. Ayudan a identificar la extensi√≥n de los datos sin incluir los valores at√≠picos.
-  **Outliers (Valores At√≠picos)**: Representan los valores que se encuentran fuera del rango de los bigotes.

El boxplot es especialmente √∫til para comparar la distribuci√≥n de los datos entre diferentes grupos o categor√≠as y para identificar r√°pidamente anomal√≠as o valores extremos en un conjunto de datos.

El boxplot es una herramienta poderosa para el an√°lisis exploratorio de datos, permitiendo a los analistas visualizar r√°pidamente la distribuci√≥n de los datos y detectar valores at√≠picos. Su capacidad para resumir la informaci√≥n clave de un conjunto de datos en un formato visual compacto lo hace ideal para comparar distribuciones entre diferentes grupos o categor√≠as.

Al integrar boxplots en el an√°lisis de datos, se puede obtener una comprensi√≥n m√°s profunda de las caracter√≠sticas subyacentes de los datos, facilitando la toma de decisiones informadas y la identificaci√≥n de patrones significativos

```python
np.random.seed(0) # Semilla 0 para que al volver a ejecutar se mantenga ese random obtenido
ages = [np.random.normal(30, 5, 100),
        np.random.normal(40, 5, 100),
        np.random.normal(40, 5, 100)]
print(ages)
```

```python
# Graficamos un Boxplot
plt.boxplot(ages, patch_artist=True, notch=True, vert=True, labels=['grupo 1', 'Grupo 2', 'Grupo 3'])
plt.title('Distribuci√≥n de edades por grupo')
plt.xlabel('Grupo')
plt.ylabel('Edad (a√±os)')
plt.show()

```

#### Diferencias claves entre estos dos tipos de gr√°ficos

Diferencias clave:

- **Histograma**: Muestra c√≥mo se distribuyen los valores en diferentes intervalos.

- **Boxplot**: Resalta los cuartiles, la mediana y posibles valores at√≠picos.

Ambos son √∫tiles dependiendo de lo que desees analizar sobre la distribuci√≥n de tus datos.
