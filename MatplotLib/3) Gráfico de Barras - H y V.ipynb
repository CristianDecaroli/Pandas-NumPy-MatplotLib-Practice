### Gr√°fico de Barras Verticales

````python
import matplotlib.pyplot as plt
````

````python
plt.figure(figsize=(8,6))

categories = ['Producto A', 'Producto B', 'Producto C']
sales = [120, 150, 90]

# Creaci√≥n del gr√°fico de barras verticales
plt.bar(categories, sales, color='skyblue', label='Ventas Mensuales') # M√©todo para crear el gr√°fico de barras

# Anotaci√≥n con flecha destacar un punto espec√≠fico
plt.annotate('M√°ximo de ventas', xy=('Producto B', 150), xytext=('Producto C', 160),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.title('Ventas de Productos en un Mes')
plt.xlabel('Productos')
plt.ylabel('Ventas')

plt.legend()
plt.show()
````

#### Gr√°fico de Pastel

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

products = ['Producto A', 'Producto B', 'Producto C']
market_share = [50, 35, 15]

# Creaci√≥n del gr√°fico/diagrama de pastel
plt.pie(market_share, labels=products, autopct='%1.1f%%', startangle=140, colors=['gold','lightcoral', 'orange'])

plt.title('Participaci√≥n del Producto en el Mercado')

# Asegurar que el gr√°fico sea un c√≠rculo
plt.axis('equal')

plt.show()
````

En la visualizaci√≥n de datos, los gr√°ficos de pastel son una herramienta eficaz para representar proporciones dentro de un conjunto de datos.

Una caracter√≠stica importante de estos gr√°ficos es la capacidad de personalizar c√≥mo se muestran los porcentajes de cada segmento. A continuaci√≥n, exploraremos las diferentes formas de formatear los porcentajes en un gr√°fico de pastel utilizando el par√°metro **autopct** de la funci√≥n **plt.pie()** en matplotlib.

Veremos c√≥mo utilizar cadenas de formato para ajustar la precisi√≥n decimal y a√±adir texto personalizado, as√≠ como implementar funciones para un control m√°s detallado sobre la presentaci√≥n de los datos.

1. **Cadenas de Formato (autopct)**:
- **'%1.1f%%'**: Muestra el porcentaje con un decimal y el s√≠mbolo %. Ejemplo: 35.7%.
- **'%1.0f%%'**: Muestra el porcentaje como un n√∫mero entero sin decimales. Ejemplo: 36%.
- **'%1.2f%%'**: Muestra el porcentaje con dos decimales. Ejemplo: 35.68%.
- **'%2.0f%%'**: Similar a '%1.0f%%', pero reserva m√°s espacio (dos d√≠gitos enteros). Ejemplo: 35% se ver√≠a como 35%, pero 5% se ver√≠a como 5%.
- **'%1.1f'**: Muestra el porcentaje con un decimal, pero sin el s√≠mbolo %. Ejemplo: 35.7.
- **'%1.1f units'**: Personaliza el texto. En este caso, mostrar√≠a el porcentaje seguido de la palabra ‚Äúunits‚Äù. Ejemplo: 35.7 units.

2. **Funciones (autopct)**:
En lugar de una cadena de formato, puedes pasar una funci√≥n personalizada a autopct que reciba el valor del porcentaje y devuelva una cadena de texto formateada:
```py
    def custom_autopct(pct):
        return f'{pct:.2f}% ({int(round(pct/100.*sum(participacion)))} unidades)'

    plt.pie(participacion, labels=productos, autopct=custom_autopct)
    # Esta funci√≥n mostrar√≠a tanto el porcentaje con dos decimales como el valor absoluto correspondiente en unidades.
```

**Ejemplo de Uso**:
- Mostrar porcentajes con un decimal y s√≠mbolo %: **autopct='%1.1f%%'**
- Mostrar porcentajes como n√∫meros enteros: **autopct='%1.0f%%'**
- Mostrar porcentaje y a√±adir texto personalizado: **autopct='%1.1f unidades vendidas'**
- Mostrar porcentajes con formato personalizado mediante una funci√≥n: **autopct=custom_autopct**.

Estos valores permiten personalizar c√≥mo se presentan los datos en tu gr√°fico de pastel, adapt√°ndolo a tus necesidades espec√≠ficas.

Personalizar la presentaci√≥n de los porcentajes en un gr√°fico de pastel no solo mejora la est√©tica del gr√°fico, sino que tambi√©n puede hacer que la informaci√≥n sea m√°s comprensible para el p√∫blico.

Con estas herramientas, puedes crear gr√°ficos que no solo sean informativos, sino tambi√©n visualmente atractivos.
