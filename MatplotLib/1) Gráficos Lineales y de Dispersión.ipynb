#### Gráfico de Líneas

```python
import numpy as np
import matplotlib.pyplot as plt

month = np.array(['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'])
sales = np.array([20, 25, 30, 28, 35])

# Configurar el tamaño del gráfico
plt.figure(figsize=(8,6))

# Crear el gráfico
plt.plot(month, sales, marker='o', color='blue')

# Título del gráfico
plt.title('Ventas mensuales de un producto')

# Etiqueta del eje X del gráfico
plt.xlabel('Meses')

# Etiqueta del eje Y del gráfico
plt.xlabel('Ventas en miles de unidades')

# Mostrar el gráfico
plt.show()
```

#### Gráfico de Dispersión

```python
import matplotlib.pyplot as plt

# Configurar el tamaño del gráfico
plt.figure(figsize=(8,6))

hours = [2, 3, 4, 5, 6, 7, 8]
exam_score_student_1 = [55, 60, 65, 70, 75, 80, 85]
exam_score_student_2 = [50, 58, 63, 69, 74, 78, 83]

# Crear gráfico de dispersión de dos estudiantes
plt.scatter(hours, exam_score_student_1, marker='o', color='green', label='Estudiante 1') # Método para dispersión
plt.scatter(hours, exam_score_student_2, marker='s', color='red', label='Estudiante 2') # Método para dispersión


plt.title('Relación horas estudiadas y el puntaje')

plt.xlabel('Horas')

plt.xlabel('Puntaje')

plt.show()
```
