#### SERIES Y MANEJO DE FECHAS

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.dates import DateFormatter # Se importa específicamente formato de fechas desde el módulo de fechas.
```

```python
dates = pd.date_range(start='2023-01-01', periods=100) # Dataset, rango de fechas que inician desde la fecha aclarada, 100 fechas.

values = np.random.rand(100).cumsum() # Genera con random 100 números y aplica la suma acumulada

data = pd.DataFrame=({'Date': dates, 'Values': values})
```

```python
# Crear gráfico de líneas
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(data['Date'], data['Values'], color='green')  # Cambiar 'Value' a 'Values'

plt.xticks(rotation=45)

plt.title('Serie de tiempo con formato en las fechas')
plt.xlabel('Date')
plt.ylabel('Values')  # También corregir la etiqueta del eje Y

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generar datos
dates = pd.date_range(start='2023-01-01', periods=12, freq='M')
sales = np.random.randint(1000, 5000, size=12)

# Crear DataFrame correctamente
sales_data = pd.DataFrame = ({'Date': dates, 'Sales': sales})

# Crear gráfico
plt.plot(sales_data['Date'], sales_data['Sales'], marker='o', linestyle='-', label='Ventas Mensuales')

plt.gca().xaxis.set_major_formatter(DateFormatter('%b %Y')) # Formato para fecha

plt.xticks(rotation=45)

plt.title('Análisis de Ventas Mensuales')
plt.xlabel('Date')
plt.ylabel('Sales')

plt.legend()
plt.tight_layout()
plt.show()
```
