## Guardar y visualizar

```python
import numpy as np
```

```python
sales = np.random.randint(100, 301, size = 12) #Genera un array de numeros aleatorios enteros, array de 12 numeros entre 100 y 300, se almacena en la var sales
print(sales)
np.save('sales.npy', sales) # Guarda el array en un archivo llamado sales.npy, extensión eficiente para guardar datos numéricos de NumPy que luego puede ser cargado con 'np.load()'
```

```python
loaded_sales = np.load('sales.npy') # Se carga el archivo con np.load()
print(loaded_sales)
```

```python
# Este código guarda en un archivo.csv el array sales, al delimitarlo por coma se guarda en extensión .csv
np.savetxt('sales.csv', sales, delimiter=',')
```

```python
loaded_sales_csv = np.loadtxt('sales.csv', delimiter=',')
print(loaded_sales_csv)
```

```python
import numpy as np
import matplotlib.pyplot as plt

months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(10,5))
plt.bar(months,sales, color = 'b', label = 'Ventas generadas')
plt.title('Ventas por mes')
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.xticks(rotation = 45)
plt.show()
```
