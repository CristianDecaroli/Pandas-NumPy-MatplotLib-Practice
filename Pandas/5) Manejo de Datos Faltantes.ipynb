```python
import pandas as pd
```

```python
retail_data = pd.read_csv('online_retail.csv', encoding='latin1')
```

1. Con el método ***isna()*** podemos consultar si los datos están disponibles o si son nulos. Si devuelve True, es porque NO están disponibles, es decir, nulos.

```python
missing_data = retail_data.isna()
missing_data
```

2. Si bien en el output anterior no se ven datos faltantes, hay muchas filas que aún no hemos visto. Por ende podemos contar cuantos datos faltantes tenemos en total sumándolos.

```python
missing_data_count = retail_data.isna().sum()
missing_data_count
# Nótese en el output la cantidad(suma) de datos faltantes por cada columna
```

Tenemos dos opciones cuando nos encontramos con datos faltantes:
- Eliminar las filas con datos faltantes
- Completar ceros en los campos con datos faltantes

3. Elminar las filas con datos faltantes:

```python
no_missing_rows = retail_data.dropna()
no_missing_rows # Imprime todo el df pero esta vez no tendrá las filas con datos faltantes
```

4. A continuación eliminaremos las columnas con datos faltantes. Para ello debemos especificar axis=1 en el método: ***dropna(axis=1)*** 

```python
no_missing_columns = retail_data.dropna(axis=1)
no_missing_columns
# Nótese que se elminaron 2 columnas, quedando solo 6, las cuales tenían datos faltantes...
```

5. A continuación completamos los datos faltantes con ceros. Esto nos permite mantener la cantidad original de filas y columnas, sin tener que eliminarlas y así evitar perder datos que podrían ser útiles:

```python
retail_data_fillef_zeros = retail_data.fillna(0)
# Habiendo agregado los 0, podemos visualizar el output de reailt_data_fillef_zeros o también hacer un conteo de los datos nulos por columna
retail_data_fillef_zeros_count = retail_data_fillef_zeros.isna().sum()
retail_data_fillef_zeros_count
```

6. También podemos hallar la media de los datos por columa, y agregar a esta en las celdas donde haya datos faltantes, logrando así menor disperción de los datos.

```python
# Capturo la media de la columa
mean_unit_price = retail_data['UnitPrice'].mean()
# Coloco la variable anterior como dato a agregar a la columa UnitPrice en sus celdas sin datos
retail_data_filled_mean = retail_data['UnitPrice'].fillna(mean_unit_price)
retail_data_filled_mean
```
